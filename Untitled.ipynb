```python
import pandas as pd
import glob
```

```python
# Get all CSV files in the directory
csv_files = sorted(glob.glob('df*.csv'))
dfs = []

# Loop through the files and read them into variables df_1 to df_26
for i, file in enumerate(csv_files, 1):
    df = pd.read_csv(file, index_col = '날짜')
    df.index = pd.to_datetime(df.index)
    dfs.append(df.iloc[:, 0])
    
merged_df = pd.concat(dfs, axis=1)
merged_df.columns = [f'df_{i}' for i in range(1, len(dfs) + 1)]
merged_df.isna().sum()

# df_1, df_2는 결측치가 너무 많아 제거
# 나머지는 결측이 있는 row를 제거
merged_df.drop(['df_1', 'df_2'], axis=1, inplace=True)
print(merged_df.shape)
merged_df.dropna(axis=0, how='any', inplace=True)
print(merged_df.shape)
merged_df.sort_index(inplace=True)
merged_df.columns = ['KRE', 'QQQ', 'SPYX', 'VAW', 'VDE', 'VFH', 'VGT', 'VHT', 'VIG', 'VOO', 'VOX', 'VPU', 'VTWO', 'VYM', 'XAR', 'XBI', 'XHB', 'XLE', 'XLP', 'XLRE', 'XLU', 'XLV', 'XME', 'XRT']
merged_df
```

```python
df = merged_df.loc['2016-11-07':]
```

```python
for col in df.columns:
    print(col, (100*(df.loc['2020-12-31', col] - df.loc['2016-11-07',col])/df.loc['2016-11-07',col]).round(2), '%')
```
